```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
# Cargar los datos
df = pd.read_csv("regLin.csv")

# Separo en entrenamiento y pruebas


dX_train, dX_test, dy_train, dy_test = train_test_split(df[['X']],df[['y']])



# Creo el objeto de Regresi√≥n Lineal
regr = linear_model.LinearRegression()

# Entreno el modelo usando los datos de entrenamiento
regr.fit(dX_train, dy_train)

# Hago predicciones usando los datos de prueba
dy_pred = regr.predict(dX_test)

# Coeficientes
print('Coefficientes: ', regr.coef_)
# Error cuadr√°tico medio
print("Error Cuadr√°tico Medio: %.2f"
      % mean_squared_error(dy_test, dy_pred))
# Puntaje de varianza: 1 es predicci√≥n perfecta
print('Puntaje de varianza: %.2f' % r2_score(dy_test, dy_pred))

# Graficas
plt.scatter(dX_test, dy_test,  color='black')
plt.plot(dX_test, dy_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()

```

```python
def calcError(X, y, w, w0):
        return np.mean((w0 + w*X.values- y.values)**2)
w0 = np.linspace(0,6,len(dX_test))
w1 = np.linspace(-2,3,len(dX_test))

inter, c = regr.intercept_, regr.coef_
errpred= calcError(dX_test, dy_test, c, inter) 

err0 = [calcError(dX_test, dy_test, c, w0[i]) for i in range(0,len(w0))]
err1 = [calcError(dX_test, dy_test, w1[i], inter) for i in range(0,len(w1))]

```

```python
fig=plt.figure()
plt.plot(w0,err0)
fig.suptitle('Error vs pesos parametrales de la regresion', fontsize=20)
plt.xlabel('Pesos W_o', fontsize=18)
plt.ylabel('Error', fontsize=16)

plt.scatter(inter, errpred, color='red')
plt.show()
```

```python
fig=plt.figure()
plt.plot(w1,err1)
fig.suptitle('Error vs pesos parametrales de la regresion', fontsize=20)
plt.xlabel('Pesos W_1', fontsize=18)
plt.ylabel('Error', fontsize=16)

plt.scatter(inter, errpred, color='red')
plt.show()
```

```python
import random

#Cargo los datos
df = pd.read_csv("regLin.csv")

#Divido aleatoriamente los datos disponibles para entrenar y probar el modelo
X_train, X_test, Y_train, Y_test = train_test_split(df[df.columns[0:-1]],df[df.columns[-1]], train_size=0.75)

#Normaliar los datos
X_train = (X_train - X_train.mean()) / X_test.std()
Y_train = (Y_train - Y_train.mean()) / Y_train.std()

#Calculo el peso a iterar
eta = .01
x0 = 1
rmax = 1

ws = [random.randint(0,rmax) for i in range (0,X_train.shape[1] + 1)]

#Voy graficando cada regresion con cada peso a medida que desciendo

for i in range (0,X_train.shape[0]):
    v = ws[0]*x0 + (ws[1:]*X_train.iloc[i]).sum()
    t = (Y_train.iloc[i] - v)
    ws[0] = ws[0] + x0 * t
    ws[1:] = [ws[j] + X_train.iloc[i].iloc[j-1] * t * eta for j in range(1,len(ws))]   
    xmin = (X_train).min()
    xmax = (X_train).max()
    plt.plot([xmin,xmax],[ws[0]+ws[1]*xmin,ws[0]+ws[1]*xmax], linewidth = 0.5)
    
#Grafico los datos de entrenamiento
plt.scatter(X_train,Y_train)

#Grafico la regresion optima
plt.plot([xmin,xmax],[ws[0]+ws[1]*xmin,ws[0]+ws[1]*xmax],color = 'red', linewidth = 5)

plt.xlabel('Regresor', fontsize=18)
plt.ylabel('Respuesta', fontsize=16)
plt.show() 
```
